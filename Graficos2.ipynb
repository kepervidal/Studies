```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
data2 = pd.read_csv('chicken.csv')
data2.head()
```

```python
agrupado =data2.groupby(['feed'])['weight'].sum()
agrupado
```

```python
#novo df somete para testar os filtros

teste = data2.loc[data2['feed'] == 'horsebean']
teste
```

```python
#histograma considerando somete o horsebean

srn.distplot(data2.loc[data2['feed'] == 'horsebean'].weight, hist= False).set_title('horsebean')
```

```python
#histograma com valor de casein
srn.distplot(data2.loc[data2['feed'] == 'casein'].weight).set_title('casein')
```

```python
#histograma considerando somente o valor linseed
srn.distplot(data2.loc[data2['feed'] == 'linseed'].weight).set_title('linseed')
```

```python
#impressao em grfico 2 * 3
plt.figure()
plt.subplot(3,2,1)
srn.distplot(data2.loc[data2['feed'] == 'casein'].weight).set_title('casein')
plt.subplot(3,2,2)
srn.distplot(data2.loc[data2['feed'] == 'horsebean'].weight).set_title('horsebean')
plt.subplot(3,2,3)
srn.distplot(data2.loc[data2['feed'] == 'linseed'].weight).set_title('linseed')
plt.subplot(3,2,4)
srn.distplot(data2.loc[data2['feed'] == 'meatmeal'].weight).set_title('meatmeal')
plt.subplot(3,2,5)
srn.distplot(data2.loc[data2['feed'] == 'soybean'].weight).set_title('soybean')
plt.subplot(3,2,6)
srn.distplot(data2.loc[data2['feed'] == 'sunflower'].weight).set_title('sunflower')
plt.tight_layout()
```
