```python
#Histograma
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('trees.csv')
base.shape
```

```python
#dados
base.head()
```

```python
#criando o histograma, considerando somente o segundo atributo de dados e com duas divisões (bins)
# a variavel h armazena  as faixas de valores de Heigt

h = np.histogram(base.iloc[:,1], bins= 6)
h
```

```python
#vizualição do histograma com 6 divisoes (bins)
plt.hist(base.iloc[:,1],bins=7)
plt.title('Árvores')
plt.ylabel('Frequencia')
plt.xlabel('Altura')
```

```python
#Grafico de densidade

import seaborn as srn
```

```python
#histograma com a linha de dsitrbuição de frequencia

srn.distplot(base.iloc[:,1], hist = True, kde= False,
            bins=6, color= "blue",
            hist_kws={'edgecolor':'black'})
```

```python
#densidade
srn.distplot(base.iloc[:,1], hist = False, kde= True,
            bins=6, color= "blue",
            hist_kws={'edgecolor':'black'})
```

```python
#desidade e histograma
srn.distplot(base.iloc[:,1], hist = True, kde= True,
            bins=6, color= "blue",
            hist_kws={'edgecolor':'black'})
```

```python


#grafico de dispersao
plt.scatter(base.Girth, base.Volume, color= 'blue', facecolors = 'none', marker= '*')
plt.title('Árvores')
plt.xlabel('Volume')
plt.ylabel('Circuferencia')
```

```python
#grafico de linha
plt.plot(base.Girth, base.Volume)
plt.title('Árvores')
plt.xlabel('Volume')
plt.ylabel('Circuferencia')
```

```python
srn.regplot(base.Girth, base.Volume, data= base, x_jitter = 0.3, fit_reg= False)
```

```python
#divisao de tela (subgraficos)

plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(base.Girth, base.Volume)
plt.subplot(2,2,2)
plt.scatter(base.Girth, base.Height)
plt.subplot(2,2,3)
plt.scatter(base.Height, base.Volume, marker ='*')
plt.subplot(2,2,4)
plt.hist(base.Volume)
```

```python
#boxplot
srn.boxplot(base["Volume"]).set_title('Volume')
```

```python
srn.boxplot(data = base)
```
