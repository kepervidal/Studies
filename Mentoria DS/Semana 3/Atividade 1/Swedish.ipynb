```python
# O dataset escolhiodo foi o Swedish Crime Rates

#Importando as bibliotecas necessárias para análise

import pandas as pd


```

```python
#Importando o dataset e vendo o seu tamanho

df = pd.read_csv('reported.csv')
df.shape 
```

```python
#vamos ver suas colunas
df.tail(10)
```

```python
#primeiro de tudo vamos comparar a população com o numero de crimes 


df.plot(y = 'crimes.total', x = 'Year')
df.plot(y = 'population', x ='Year' )

#percebe-se que, com o aumento da população, houve um aumento na quantidade de crimes totais.
```

```python
#agora, iremos verificar os crimes baseados em seus tipos, comparado com os crimes totais

#crimes de natureza sexual
df.plot('Year', 'rape')
df.plot('Year', 'sexual.offenses')
df.plot('Year', ['sexual.offenses', 'rape'])
df.plot('Year', ['crimes.total', 'rape', 'sexual.offenses'])

#após a plotangem, verifica-se que os crimes sexuais são uma porção muito pequena dos crimes (quase insignficativa)
```

```python
#crimes classificados como assalto/roubos

df.plot('Year',['robbery', 'burglary', 'vehicle.theft', 'house.theft', 'shop.theft', 'out.of.vehicle.theft',],xlabel= 'Year', ylabel = 'Number Of thefts/robberies')
df.plot('Year',['robbery', 'burglary', 'vehicle.theft', 'house.theft', 'shop.theft', 'out.of.vehicle.theft','crimes.total'],xlabel= 'Year', ylabel = 'Number Of thefts/robberies')


#verifica-se que, assaltos/roubos tambem representam uma parte muito pequena dos numeros totais de crimes

#além disso, destaque para roubos fora do carro e roubos (burglary), que tiveram uma crescente entre as decadas de 60-70, e uma queda após a decada de 90
```

```python
#vamos verificar os crimes relacionados a abuso de substancias
 
df.plot('Year', ['narcotics','drunk.driving'], ylabel= 'Number of crimes')
df.plot('Year', ['narcotics','drunk.driving','crimes.total'])

#observa-se que crimes por dirigir bêbado não cresceu com o passar dos anos, diferente dos crimes de trafico, que tiveram um certo aumento na ultima decada, com o surgimento desde o inicio dos anos 80
```

```python
#vamos juntar os dados baseados no seu tipo de crime, assim podemos ter uma visao melhor de quais tipos de crimes cresceram com o passar dos anos

df['Sexual Crimes'] = df['rape'] + df['sexual.offenses']
df['Theft'] = df['robbery'] + df['burglary'] + df['vehicle.theft'] + df['house.theft'] + df['shop.theft'] + df['out.of.vehicle.theft']
df['Personal Crimes'] = df['crimes.person'] + df['assault'] + df['murder']
df['Penal Code'] = df['crimes.penal.code'] + df['other.penal.crimes']
df['Substance Abuse'] = df['drunk.driving'] + df['narcotics']
```

```python
#agora vamos ver os graficos

df.plot('Year', ['Sexual Crimes','Theft', 'Personal Crimes', 'Penal Code', 'Substance Abuse' ], ylabel= 'Number Of Crimes')
```

```python
#com isso podemos perceber que em sua grande maioira, os crimes que crescerem com o passar dos anos foram os relacionados a crimes penais.
#entreanto não se sabe oque é considerado crime penal, então fica uma resposta genérica. 

```
