```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as sm
```

```python
#carregamento da base de dados

df = pd. read_csv('mt_cars.csv')
df.head()
```

```python
#exclusao da 1ra coluna

df = df.drop(['Unnamed: 0'], axis= 1)
df.head()
```

```python
#criação de X e Y
#calculo de correlação

x = df.iloc[:,2].values
y= df.iloc[:,0].values
correlacao = np.corrcoef(x,y)
correlacao
```

```python
#mudanca do x para matriz

x = x.reshape(-1,1)
```

```python
#criacao do modelo, treinamento, visualização dos coeficientes e do score do modelo

modelo = LinearRegression()
modelo.fit(x,y)
```

```python
#creiacao do modelo, utilizando a biblioteca statsmodel

modelo_ajustado = sm.ols(formula= 'mpg ~ disp', data= df)
modelo_treinado = modelo_ajustado.fit()
modelo_treinado.summary()
```

```python
#criacao de novass variaveis x1 e y1, para compara com o novo modelo
# 3 variáveis dependentes para prever mpg: cyl ~ disp ~hp

x1 = df.iloc[:, 1:4].values
x1
```

```python
y1= df.iloc[:,0].values
```

```python
modelo2 = LinearRegression()
modelo2.fit(x1,y1)
#calculando o R^2
modelo2.score(x1, y1)
```

```python
#criação do modelo ajustado com mais variaveis, Regressao Linear Multipla

modelo_ajustado2 = sm.ols(formula= ' mpg ~ cyl + disp + hp', data = df)
modelo_treinado2 = modelo_ajustado2.fit()
modelo_treinado2.summary()
```
